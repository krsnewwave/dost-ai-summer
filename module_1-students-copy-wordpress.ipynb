## Generators

```python
def gen_num_up_to(n):
    num = 0
    while num < n:
        yield num
        num +=1
```

```python
gen_num_up_to(5)
```

## List comprehension

```python
[v for v in gen_num_up_to(5)]
```

Here is a sample function. np.random.randint(a, [b]) outputs uniformly random values from $[a,b) $

```python
def remainder(n, m):
    while True:
        if n - m < 0:
            return n
        else:
            n = n - m
```

```python
remainder(10, 4)
```
